# A. Nonlinear Transformation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
from sklearn.metrics import r2_score

# Get the data
gdp_df =  pd.read_csv('china_gdp.csv')
gdp_df.head()
```

```python
# Get the x and y values
x = gdp_df['Year']
y = gdp_df['Value']

# Get slope and intercept
y_log = np.log(y)
x_log = np.log(x)
slope, intercept, rvalue, pvalue, stderr = linregress(x_log, y_log)

# Predict using the power model equation
y_pred = np.exp(intercept + slope * np.log(x))

# Compute r2
r2 = r2_score(y, y_pred)
print(f"The coefficient of determination or r squared is {r2}")
```

```python
residual = y - y_pred

# Residual plot
plt.scatter(x, residual, c='red', label='Residuals')
plt.axhline(y=0, color='blue', linestyle='--', label='Zero Residual Line')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.legend()
plt.title('Residual Plot')
plt.grid(True)
plt.show()
```
