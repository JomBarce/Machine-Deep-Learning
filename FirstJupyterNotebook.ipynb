### Graphing with Pandas and Matplotlib

#### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Get data

```python
data_url = ("https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv")
df = pd.read_csv(data_url)
```

#### Check data

```python
df.head()
```

#### Show the graph

```python
df.plot(x="Rank", y=["P25th", "Median", "P75th"])
plt.title("Earnings of Graduates in each Major")
plt.ylabel("Earnings")
```

```python
top_medians = df[df["Median"] > 60000].sort_values("Median")
top_medians.plot(x="Major", y=["P25th", "Median", "P75th"], kind="bar")
plt.title("Top Medians")
plt.ylabel("Earnings")
```
